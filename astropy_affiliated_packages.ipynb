## astroquery: online database querying

AstroQuery is a set of tools for querying astronomical web forms and databases 
([code](), [docs](http://astroquery.readthedocs.org/en/latest/)).

We'll just pick one data service here to show how it works:

* MAGPIS: The Multi-Array Galactic Plane Imaging Survey ([webpage](http://third.ucllnl.org/gps/))


```python
from astroquery import magpis
print('Available surveys: %s' % magpis.magpis.surveys)
```

```python
# Download an image with the Galactic centre in the middle and size 5 x 5 deg
# Since save=False we don't write it to disk, but get an in-memory object containing the data
fitsfile = magpis.get_magpis_image_gal(glon=0, glat=0, survey='gpsmsx', size=5 * 60, save=False)
# The download progress display doesn't work in IPython currently,
# be a bit patient, you are downloading 4 MB of data
```

```python
# What did we get?
type(fitsfile)
```

```python
# fitsfile[0] is a 1024 x 1024 pixel image
print(type(fitsfile[0]))
print(fitsfile.info())
```

```python
# Plot the image using http://aplpy.github.com/
# (astropy doesn't feature astronomy image plotting yet)
import aplpy
figure = aplpy.FITSFigure(fitsfile[0])
figure.show_colorscale()
figure.show_colorbar()
```

Obviously we didn't get a 5 x 5 deg image as requested, but approximately a 0.4 x 0.4 deg image. There should have been an error or warning.
Also the image isn't 2048 x 2048 as the `magpis.get_magpis_image_gal` docstring claims, but 1024 x 1024.

## photutils: photometry tools

The `photutils` package ([code](https://github.com/astropy/photutils/), [docs](http://photutils.readthedocs.org/)) is destined to implement functions for

* detecting sources on astronomical images
* estimating morphological parameters of those sources
  (e.g., centroid and shape parameters)
* performing photometry (both aperture and PSF)

Currently, only aperture photometry functions have been implemented.

It is possible that `photutils` will eventually be merged into `astropy` as
`astropy.photometry`.
